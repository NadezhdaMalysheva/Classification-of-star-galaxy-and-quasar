# Классификация по фотометрическим данным

Расположение данных:

> Папка, в которой находятся дополнительные данные(в ней так же должена находиться папка models с массивами numpy с фотометрическими данными для предсказания)



> class.py


> prediction.py







Сохранять массив nampy таким образом: np.save('name_data', массив)

Результат будет находиться в папке с данными с названием prediction_(вид классификации).npy

Чтобы его открыть достаточно np.load('(путь до файла)/prediction_(вид классификации).npy')

## Запускать : ! python3 [3 или больше аргумента:]

     

1.   (путь до программы)/**class.py**
2.   **вид классификации**
      *   OB_GQ
      *   OB_ST
      *   ST
3. **путь до папки с файлами**
4.   **название файла с фотометрическими данными**
5.   *формат возврата* (если есть этот аргумент, что возвращает массив вероятности принадлежностьи объектов к необходимому классу)





      
 
          
   
 


Запуск на примере тестовых данных, подготовленных ранее. (X_test_gq.npy)

```python
! python3 ./class.py OB_GQ ./input X_test_gq_dp.npy
```

В папке input появился файл prediction_GQ.npy

Проверим точность полученной классификации

```python
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
```

```python
y_test = np.load('./input/y_test_gq_dp.npy') #на этих данных будем проверять
```

```python
y_pred = np.load('./input/prediction_GQ.npy')
```

```python
print('Classification Report: \n', classification_report(y_test, y_pred))
print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
print('Accuracy score: \n', accuracy_score(y_test, y_pred))
```
